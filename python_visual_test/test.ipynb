```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.animation import FuncAnimation as anim
```

```python
x , y = [] , []
abscisses = np.arange(-10,10,10000)

for i in abscisses:
    x.append(i)
    y.append(np.sin(i)) #overkill car j'utilise numpy sur des float uniques...
    plt.scatter(x,y)
    plt.pause(0.1)
    plt.show()
```
